```python
import re

# Function to parse lines with 'param = param_value' format and filter by substring
def parse_namelist_with_substring(file_path, substring):
    param_dict = {}
    pattern = re.compile(rf'^(\w*{substring}\w*)\s*=\s*(.+)$')  # Match '*gw*' format
    with open(file_path, 'r') as file:
        for line in file:
            # Strip whitespace and ignore comment lines
            line = line.strip()
            if line.startswith("#") or not line:
                continue
            #print(line)
            # Match lines containing the specific substring
            match = pattern.match(line)
            if match:
                param, value = match.groups()
                # Convert the value to int, float, or leave as string
                try:
                    value = eval(value)  # Evaluate literals safely
                except:
                    value = value.strip('"').strip("'")  # Handle strings explicitly
                param_dict[param] = value
    return param_dict


# Example usage
file_path = "example.txt"
exp="c64_48_gwR2_ne30pg3_FMTHIST_x05" 
file_path = f"/glade/derecho/scratch/juliob/{exp}/run/atm_in"

substring = "effgw"  # Specify the substring
params = parse_namelist_with_substring(file_path, substring)
print(params)

```

```python
def parse_namelist_without_re(file_path, substring):
    param_dict = {}
    with open(file_path, 'r') as file:
        for line in file:
            # Strip whitespace and ignore comment lines
            line = line.strip()
            if line.startswith("#") or not line:
                continue

            # Check if the line contains '=' and the substring
            if '=' in line and substring in line:
                # Split into parameter and value parts
                param, value = line.split('=', 1)
                param = param.strip()
                value = value.strip()

                # Convert value to int, float, or leave as string
                try:
                    value = eval(value)  # Evaluate literals safely
                except:
                    value = value.strip('"').strip("'")  # Handle strings explicitly

                # Add to the dictionary if the parameter contains the substring
                if substring in param:
                    param_dict[param] = value
    return param_dict

# Example usage
file_path = "example.txt"
substring = "gw"  # Specify the substring
params = parse_namelist_without_re(file_path, substring)
print(params)
```
