```python
import re
import yaml
import glob
import os


# Function to parse lines with 'param = param_value' format and filter by substring
def parse_namelist_with_substring(file_path, substring):
    param_dict = {}
    pattern = re.compile(rf'^(\w*{substring}\w*)\s*=\s*(.+)$')  # Match '*gw*' format
    with open(file_path, 'r') as file:
        for line in file:
            # Strip whitespace and ignore comment lines
            line = line.strip()
            if line.startswith("#") or not line:
                continue
            #print(line)
            # Match lines containing the specific substring
            match = pattern.match(line)
            if match:
                param, value = match.groups()
                # Convert the value to int, float, or leave as string
                try:
                    value = eval(value)  # Evaluate literals safely
                except:
                    value = value.strip('"').strip("'")  # Handle strings explicitly
                param_dict[param] = value
    return param_dict

# File path to the existing YAML
#yaml_file = "GWppe2_Copy.yaml"
yaml_file = "GWppe2.yaml"
#yaml_file = "GW_super_ppe.yaml"
#yaml_file = "GW_Xtreme_ppe.yaml"

# Read the YAML file
try:
    with open(yaml_file, 'r') as file:
        data = yaml.safe_load(file)  # Load the existing YAML into a dictionary
except FileNotFoundError:
    print(f"File '{yaml_file}' not found. Starting with a new structure.")
    data = {}


# Loop through all top-level keys
for key, value in data.items():
    print(f"Key: {key} ")

    x=key  #'L'

    file_path=f"{data[x]['caseroot']}{data[x]['name']}/CaseDocs/atm_in"
    print(file_path)
    if not os.path.exists(file_path):
        print(f"File not found: {file_path}. Skipping {key}.")
        continue  # Skip to the next key

    if 'gw_parameters' not in data[x]:
        data[x]['gw_parameters'] = {}


    substring = "effgw"  # Specify the substring
    params = parse_namelist_with_substring(file_path, substring)
    print(params)
    # Add the new dictionary under Parameters
    data[x]['gw_parameters'].update(params)
    
    substring = "taubgnd"  # Specify the substring
    params = parse_namelist_with_substring(file_path, substring)
    print(params)
    # Add the new dictionary under Parameters
    data[x]['gw_parameters'].update(params)
    
    substring = "poopoo"  # Specify the substring
    params = parse_namelist_with_substring(file_path, substring)
    print(params)
    # Add the new dictionary under Parameters
    data[x]['gw_parameters'].update(params)
    
    substring = "use_gw"  # Specify the substring
    params = parse_namelist_with_substring(file_path, substring)
    print(params)
    # Add the new dictionary under Parameters
    data[x]['gw_parameters'].update(params)
    
    substring = "alpha_gw"  # Specify the substring
    params = parse_namelist_with_substring(file_path, substring)
    print(params)
    # Add the new dictionary under Parameters
    data[x]['gw_parameters'].update(params)
    
    substring = "movmtn_"  # Specify the substring
    params = parse_namelist_with_substring(file_path, substring)
    print(params)
    data[x]['gw_parameters'].update(params)
    
    ###################################
    # Now look for SourceMods 
    ###################################
    if 'SourceMods' not in data[x]:
        data[x]['SourceMods'] = {}
    
    # Specify the directory to search
    file_path =f"{data[x]['caseroot']}{data[x]['name']}/SourceMods/"
    
    # Find all .F90 files
    f90_files = glob.glob(f"{file_path}/*/*.F90")
    
    # Extract only the file names
    file_names = [os.path.basename(file) for file in f90_files]
    print("Found .F90 file names:")
    for name in file_names:
        print(name)
    
    
    params={'files':file_names}
    data[x]['SourceMods'].update(params)
    


# Write the updated structure back to the YAML file
with open(yaml_file, 'w') as file:
    yaml.dump(data, file, default_flow_style=False)
```
