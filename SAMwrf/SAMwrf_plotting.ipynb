```python
phis=ds['PHIS'][0,:].values/9.8
phisx,Xlon,Ylat=a.c_o_xy(idata=phis,lon=lon,lat=lat,dx=0.02,dy=0.02)

v=ds['V'][0,:,:].values

```

```python
cmap=plt.cm.terrain
fig = plt.figure(figsize=(20, 9))

clevels=np.linspace(-1000,5000,31)

ax1 = fig.add_subplot(1, 1, 1 , projection=ccrs.PlateCarree(central_longitude=0))
cf = ax1.contourf(Xlon, Ylat, phisx,  levels=clevels, cmap=cmap  , transform=ccrs.PlateCarree())

ax1.coastlines(resolution='110m')
    
gl=ax1.gridlines(linestyle=':', color='black',draw_labels=True)  #, ylabels_right=None)

ax1.set_extent([270., 340, -60., 20.], crs= ccrs.PlateCarree() )


gl.right_labels=False
gl.top_labels=False
        
plt.colorbar(cf , ax=ax1 , shrink=0.8)
```

```python
ps=ds['PS'][0,:].values
hyam=ds['hyam'].values
hybm=ds['hybm'].values

p3=a.press(PS=ps,hybm=hybm,hyam=hyam )
print(np.shape(p3))
```

```python

dimo = np.shape(u)
print( dimo[0], dimo[1] )
```

```python

Plev=np.asarray( [1000., 950. , 850.,700.,500.,200.] )*100.
nPlev=len(Plev)
print(nPlev) 
```

```python
uP=np.zeros( ( nPlev , dimo[1]) )
for icol in np.arange( dimo[1] ):
    fint=intr.interp1d( x=p3[:,icol],y=u[:,icol] , bounds_error=False, fill_value=0. )
    uP[ :, icol ] = fint( Plev )
```

```python
vP=np.zeros( (nPlev, dimo[1]) )
for icol in np.arange( dimo[1] ):
    fint=intr.interp1d( x=p3[:,icol],y=v[:,icol] , bounds_error=False, fill_value=0. )
    vP[ :, icol ] = fint( Plev )
```

```python
print(np.shape(lat))
g10=np.linspace(0,1_167_000.,num=1_000,dtype=np.int64)
print(np.shape(g10))
```

```python
cmap=plt.cm.terrain
fig = plt.figure(figsize=(20, 9))

clevels=np.linspace(-1000,5000,31)

ax1 = fig.add_subplot(1, 1, 1 , projection=ccrs.PlateCarree(central_longitude=0))
cf = ax1.contourf(Xlon, Ylat, phisx,  levels=clevels, cmap=cmap  , transform=ccrs.PlateCarree())


ax1.coastlines(resolution='110m')
    
gl=ax1.gridlines(linestyle=':', color='black',draw_labels=True)  #, ylabels_right=None)

L=1
vecs=50
ax1.quiver(lon[g10], lat[g10], uP[L,g10],vP[L,g10] , scale=vecs, scale_units='inches',transform=ccrs.PlateCarree())
ax1.set_extent([270., 340, -60., 20.], crs= ccrs.PlateCarree() )


gl.right_labels=False
gl.top_labels=False
        
plt.colorbar(cf , ax=ax1 , shrink=0.8)


#fig = plt.figure(figsize=(20, 9))
#ax = plt.plot(projection=ccrs.PlateCarree(central_longitude=0))
#cf = plt.contourf(Xlon, Ylat, phisx,  levels=31, cmap=cmap , transform=ccrs.PlateCarree())
```
